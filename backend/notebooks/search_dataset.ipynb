```python
import fitz

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    document = fitz.open(pdf_path)

    # Extract text from each page
    doc_blocks = []
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        page_blocks = page.get_text("blocks")
        for block in page_blocks:
            doc_blocks.append(block[4].strip())

    return doc_blocks

doc_blocks = extract_text_from_pdf("../data/LayoutLM.pdf")
```

```python
all_text = " ".join(doc_blocks)
```

```python
all_words = all_text.split()
```

```python
import arxiv
```

```python
arxiv_id = "2301.00001"  # Replace this with your desired paper's arXiv ID

# Search for the paper using its ID
search = arxiv.Search(id_list=[arxiv_id])
```

```python
url = list(search.results())[0].pdf_url
```

```python
import requests
from io import BytesIO
response = requests.get(url)
```

```python
pdf_content = BytesIO(response.content)
pdf_content.getvalue()
```

```python
Document = fitz.Document(stream=pdf_content, filetype="pdf")
```
